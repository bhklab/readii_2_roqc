```python

import numpy as np
import pandas as pd
from damply import dirs
```

```python
dataset = "TCGA-KIRC"

mit_index = pd.read_csv(dirs.PROCDATA / f"TCIA_{dataset}" / "images" / f"mit_{dataset}" / f"mit_{dataset}_index-simple.csv")

```

```python
seg_list = pd.read_csv(dirs.RAWDATA / f"TCIA_{dataset}" / "images" / "TCGA-KIRC_segmentations_Feng" / "list.csv", index_col=0)

seg_list.rename(columns={'Patient': 'PatientID',
                         'SeriesInstanceUID': 'ReferencedSeriesUID',
                         'path': 'ImageID'}, inplace=True)

seg_list['SeriesInstanceUID'] = seg_list['mask'].str.removesuffix('_.nii.gz').str.split('_', n=2).str[2]

seg_list['filepath'] = "TCGA-KIRC_segmentations_Feng/" + seg_list['mask']
seg_list['Modality'] = 'NIFTI'

mit_seg_list = seg_list[['filepath', 'ImageID', 'PatientID', 'Modality', 'SeriesInstanceUID', 'ReferencedSeriesUID']].copy()
```
